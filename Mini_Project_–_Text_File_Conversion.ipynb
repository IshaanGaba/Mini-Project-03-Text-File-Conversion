```python
import seaborn as sns
```

```python
x=[i*100 for i in range(2,11,2)]
y1=[12,20,34,45,55]
y2=[15,25,36,50,60]
y3=[18,30,40,55,70]
y4=[20,35,45,60,80]
y5=[25,40,53,75,100]
s=sns.lineplot(x=x,y=y1,label='C',linestyle='-',marker="o")
sns.lineplot(x=x,y=y2,label='C++',linestyle='-',marker="s")
sns.lineplot(x=x,y=y3,label='JAVA',linestyle='-',marker="^")
sns.lineplot(x=x,y=y4,label='R',linestyle='-',marker="*")
sns.lineplot(x=x,y=y5,label='PYTHON',linestyle='-',marker="d")
s.set_title('Performance Comparison')
s.set_xlabel('File Size(MB)')
s.set_ylabel('Time Taken (Sec)')
```
